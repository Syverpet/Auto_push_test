```python
import pandas as pd

from googleapiclient.discovery import build
from google.oauth2 import service_account

# If modifying these scopes, delete the file token.json.
SCOPES = ['https://www.googleapis.com/auth/spreadsheets']

"""Shows basic usage of the Sheets API.
Prints values from a sample spreadsheet.    
"""
creds = None
# The file token.json stores the user's access and refresh tokens, and is
# created automatically when the authorization flow completes for the first
# time.

creds = service_account.Credentials.from_service_account_file('C:/Users/Laptop/Documents/Jobb/Geo start-up/Projects/Google_Sheets_API/test/KEY_sheets_API_test.json', scopes=SCOPES)
# If there are no (valid) credentials available, let the user log in.

SAMPLE_SPREADSHEET_ID = '1qvUguizcGvMqC2u9CmO2PJMJ0IvOm91xn1nj3F3eov8'


service = build('sheets', 'v4', credentials=creds)

# Call the Sheets API
sheet = service.spreadsheets()
result = sheet.values().get(spreadsheetId=SAMPLE_SPREADSHEET_ID,
                            range="Vaccinations!A1:G500").execute()
values = result.get('values', [])

print(result)

# make dataframe from sheets values
df = pd.DataFrame(values[1:], columns = values[0])

df.to_csv('C:/Users/Laptop/Documents/Jobb/Geo start-up/Projects/Google_Sheets_API/test/sheets_to_csv_test.csv', index=False)

df
```
